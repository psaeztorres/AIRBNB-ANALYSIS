Now the subset is clean it is time to look for the answer to question 1. As we will do queries very symilar for three 
different options, the most efficient way is to create a function to do so.

```python
def get_top_3_neighbours (df, room_type):
    '''this function takes a dataframe and a room_type and retunrsa df with the 3 NYC neighbourhoods with more Rb&b offer
    of that room_type and the quantity in each neighbourhood
    
    Parameters:
    df (DataFrame) : Dataframe
    room_type (str) : room_type value 
    '''
    
    df1 = df[(df['room_type'] == room_type)].value_counts().head(3)
    return pd.DataFrame(df1)
```

### QUESTION 2- Neighbourhood with more available bedrooms, does this data affect the average renting price of the neighbourhood?

```python
data.sample()
```

The data we need to answer this question is: neighbourhood, bedrooms, price. Let's create the subset and check for NaN

```python
q2_subset = data[['neighbourhood','bedrooms','price']]
```

```python
q2_subset.isna().sum()
```

again just in neighbourhoods so we will drop them

```python
q2_subset.dropna(inplace=True)
```

In order to answer this question we must create a new column 'neigh_price' that will hold the average price for each neighbourhood so we will be able to have the data we need to anser this question: number of beds and average price in each neighbourhood. Column 'price' is then no longer needed and we drop it

```python
q2_subset['neigh_price']=q2_subset.groupby('neighbourhood')['price'].transform('mean').round(3)
q2_subset.drop(columns ='price', inplace = True)
```

```python
q2_subset.head(3)
```

In order to know which is the neighbourhood with highest number of bedrooms we just have to sort the dataframe by this column. But to answer if this fact affects the price the best way to see it (although we can see it in the result ordered df) will be a visual one: a lineplot

```python
q2_neigh = pd.DataFrame(q2_subset.groupby('neighbourhood').agg({'bedrooms':'sum','neigh_price':'first'}).sort_values(by='bedrooms',ascending=False))
q2_neigh.head(10)
```

```python
plt.figure(figsize=(12,4))
sns.lineplot(data=q2_neigh, x='neigh_price',y='bedrooms')
plt.xticks(rotation=45)
plt.axvline(x=111.548, color='red', linestyle='--', label='avg_price Brooklyn')
plt.legend()
plt.title('avg price/ neighbourhood vs bedrooms avilable/neighbourhood');
```

<u>ANSWER TO QUESTION 2</u> 

The neighbourhood with highest bedrooms offered is Brooklyn and this is not affecting the average price in this zone.

### QUESTION 3- Does the total number of amenities affect the final price?

```python
data.sample()
```

To get the answer to this question we will carry on a similar process to question 2 but, this time, the new column we will have to create is the one containing how many amenities are included in each row ('num_amenit').
In order to do this we will, for each value in 'amenities' column, split it by ',' and then count the elements by using the len function. 

```python
q3_subset = data[['amenities', 'price']]
```

```python
q3_subset['num_amenit'] = data['amenities'].apply(lambda x: len(x.split(',')))
q3_subset.drop(columns = 'amenities', inplace = True)
q3_subset.sample(3)
```

In order to see if there is a correlation between the number of amenities offered and the price a lineplot can be very usefull

```python
plt.figure(figsize=(12,4))
sns.lineplot(data=q3_subset, x='num_amenit',y='price')
plt.xlabel('number of amenities offered')
plt.title('price vs number of amenities offered');
```

<u>ANSWER TO QUESTION 3</u> 
As is often the case with any analysis, the final price is influenced by numerous factors. Therefore, a plot featuring only one variable is inevitably affected by other variables. Keeping this in mind, the plot above reveals that, for the number of amenities ranging approximately between 5 and 35, there is a discernible correlation between the number of amenities and the price. Beyond or below this range, it appears that the influence of other factors hinders our ability to observe whether a correlation exists or not.


### QUESTION 4- Which is the percentaje of rentals offering each listed amenity?

In order to answer this question we will have to make a list with all the amenities offered and, afterwards, check how many rentals offer each one.

```python
data.sample()
```

```python
q4_subset=data[['id','amenities']]
```

```python
amenities_str=','.join(q4_subset['amenities']) #we create a unique string with all the values from 'amenities' column
```

```python
lista = amenities_str.replace('"','').replace('{','').replace('}','').split(',') 
#we clean the previous string removing '{','}', and '""' and finalyy splitting by ','
lista
```

```python
amenities_unique=list(set(lista))
amenities_unique
```

```python
#as we have detected '' in the list, we must remove this elements
amenities_unique = [i for i in amenities_unique if i != '']
```

```python
amenities_unique
```

Finally, we must iterate through the dataframe to determine the frequency of each element in 'amenities_unique.' To accomplish this, we will create a function ('frequency_in_df') that returns the frequency of each element in a list appearing in a specific column of the dataframe. This function can be utilized for other queries if necessary.

```python
def frquency_in_df (lista, df, column):
    '''This function takes a list, a column name and a dataframe, and returns a dataframe showing the frequency for each 
    element in the list in the specified column of the dataframe.
    
    Parameters:
    lista (list): List of elements
    df (DataFrame): Dataframe
    column (str): Name of the column in the dataframe
    
    Returns:
    DataFrame: A dataframe showing the frequency of each element in the list in the specified column
    '''
    data = []
    for i in lista:
        rows_with_i=df[column].apply(lambda x : i in x).sum()
        percentaje = (rows_with_i / len(df)) * 100
        data.append({
            "amenity": i,
            "frequency": percentaje
            })
        final_df = pd.DataFrame(data).sort_values('frequency', ascending=False).replace('',np.nan).dropna()
    return final_df
```

<u>ANSWER TO QUESTION 4</u> 
The answer to this question will be just executing the function frequency_in_df using the parameters:

    - lista = amenities_unique that we have just created 
    - df = q4_subset
    - column = 'amenities'


```python
amenities_df =frquency_in_df (amenities_unique, q4_subset, 'amenities')
amenities_df.head(15)
```
